```python
# groupby
company=pd.read_csv('Fortune500.csv',engine="python") 
# engine used for solving encoding error
company.head(2)
```

```python
sect=company.groupby('Sector')
sect
```

```python
len(sect)  #  total 21 groups created
```

```python
sect.size().sort_values(ascending=False)
# Sector is index and other column is value
```

```python
sect.first()  # shows top company in each sector
```

```python
sect.last() #  last company in each sector 
```

```python
sect.groups  
# returns dictionary key as sector and values of everycompany 
#representing that sector
```

```python
sect.get_group('Energy')  # displays particular company
```

```python
sect.sum()
```

```python
sect.mean()
```

```python
sect['Revenues'].mean().sort_values(ascending=False)
```
